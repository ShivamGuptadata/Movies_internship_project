## Task 2: Cleaning the Data

-  ### Subtask 2.1: Inspect Null values

Number of Null values in all the columns and rows. Also, the percentage of Null values in each column. Round off the percentages upto two decimal places.

```python
#code for column-wise null count
value_na=movies.isna()
null_v=[]
for i in np.arange(len(col)):
    a=col[i]
    a1=0
    for j in np.arange(len(value_na[a])):
        a2=int(value_na[a][j])
        a1=a1+a2
    null_v=np.append(null_v,a1)
```

```python
pd.DataFrame([col,null_v]).T          #missing section found
```

```python
# code for column-wise null percentages 
null_per=round((movies.isnull().sum()/len(movies.index))*100,2)
null_per
```

```python
# row-wise null count
movies.isnull().sum(axis=1)
```

```python
# code for calculating decade
cols = ["movie_title", "director_name","imdb_score", "title_year","genres_1","genres_2"]
condition0 = df_movies["title_year"].astype(int) < 2020

df_by_decade = (df_movies
         [cols]
         [condition0]
        )

df_by_decade["decade"] = df_by_decade['title_year'].astype(int)//10*10

df_by_decade.head()
```

```python
# code for creating the data frame df_by_decade 
df_by_decade=df_by_decade.sort_values(by='decade',ascending=True)
df_by_decade
```

```python
# code for plotting number of voted users vs decade
plt.figure(figsize=(10,8))
plt.scatter(df_movies['num_voted_users'],df_by_decade['decade'])
plt.title("Plotting number of voted users vs decade")
plt.xlabel("Voted_users")
plt.ylabel("Decade")
plt.Figure(figsize=(20,15))
plt.show()
```
